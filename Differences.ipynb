# Differences (Discrete derivative).
Rumus time series :

yt = xt + 1 - xt 

Contoh implementasi kedalam python :

```python
import numpy as np

x = [1,3,48,55,66]

np.diff(x) # .diff ADALAH FUNGSI ATAU METHOD UNTUK MENCARI NILAI DIFFERENCES (SELISIH)

# DERAJAT 2 DARI VARIABEL x (HASIL DARI DERAJAT 1 DIATAS DICARI LAGI NILAI DIFFERENCES NYA) :
np.diff(x,2)
```

Secara Grafis :

```python
import numpy as np
from matplotlib import pyplot as plt

""" 
f(x) = x**2 + 1
d(f(x)) = 2x

Slope = BENTUK TURUNAN ITU SENDIRI DIMANA PERUBAHAN f(x) / PERUBAHAN x, MENJADI diff(f(x)) / diff(x)  
"""

x = np.linspace(-2,5,200) # GENERATE DATA POINT SEBANYAK 200 DATA DENGAN NILAI -2 SAMPAI 5
f = x**2 + 1 # FORMULA
dx = np.diff(x) # PERUBAHAN ATAU NILAI diff DARI x
df = np.diff(f) # PERUBAHAN ATAU NILAI diff DARI f
```

Bentuk plot dari variabel x dan f(x) :

```python
plt.plot(x,f)
plt.grid()
plt.show()
```

Bentuk plot dari variabel x dan hasil nilai differences f dan x :

```python
plt.plot(x[1:],df/dx) # df / dx ADALAH SEBUAH IMPLEMENTASI DARI FUNGSI DERIVATIVE DARI SEBUAH VARIABEL x
plt.grid()
plt.show()
```

Gabungan dari f(x) dan nilai diff :

```python
plt.plot(x,f)
plt.plot(x[1:], df/dx)
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.legend(['f','df/dx'], loc='best')
plt.grid()
plt.show()
```

Kesimpulan :

f(x) = x**2 + 1

jika x = 3, maka f = 10

jika x = 1 maka f = 2

slope = perubahan d(f) / perubahan d(x) = (10-2) / (3-1) = 8/2 = 4

hasilnya adalah = 4

artinya nilai slope 4 adalah nilai setiap penambahan 2 nilai x

atau nilai slope 2 adalah nilai setiap penambahan 1 nilai x
